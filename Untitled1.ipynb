```python
import pandas as pd
```

```python
dfFases = pd.read_csv("fases-con-nombre.csv")
```

```python
gb = dfFases.groupby("Grabacion")
```

```python
dfFrame = gb.get_group("23-07-2019")
```

```python
dfFrame["video_name"] = dfFrame.video_name.apply(lambda x : x.lower())
```

```python
for x in dfVideo_seg:
    try:
        files = {'video': open(x, 'rb')}
        url = "http://localhost:5000/splitVideo"
        r = requests.post(url,files=files)
        respuesta = r.json()
        respuesta['name'] = x
        list_json.append(respuesta)
    except Exception as e:
        print(str(e))
        r.close()
```

```python
len(list_json)
```

```python
dfVideo_seg.shape
```

```python
dfFinal = pd.DataFrame(list_json)
```

```python
dfFinal.name = dfFinal.name.apply(lambda x : x.replace("../../video_seg/",""))
```

```python
for i in range(0 , 3):
    dfFinal[f'fase-{i}'] = dfFinal[f'fase-{i}'].apply(lambda x:x.replace('image_00',""))
```

```python
for i in range(0 , 3):
    dfFinal[f'fase-{i}'] = dfFinal[f'fase-{i}'].apply(lambda x:x.replace('.jpg',""))
```

```python
for i in range(0 , 3):
    dfFinal[f'fase-{i}'] = dfFinal[f'fase-{i}'].astype(int)
```

```python
dfFinal.to_csv("salida-fases.csv")
```
